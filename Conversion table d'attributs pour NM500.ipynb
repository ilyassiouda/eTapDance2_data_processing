```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 10}) # pour augmenter la taille de caractère dans les graphiques

from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('etapdance2/dataset_cut_ilyas_math_features_float.txt')

dancer = "Ilyas"

step_type_col = dataset[['step type']]

means_col = dataset[['Gyro0X mean', 'Gyro0Y mean', 'Gyro0Z mean',
                                 'Accel0X mean','Accel0Y mean','Accel0Z mean',
                                 'Gyro1X mean', 'Gyro1Y mean','Gyro1Z mean',
                                 'Accel1X mean', 'Accel1Y mean','Accel1Z mean']]
                         
stds_col = dataset[['Gyro0X std', 'Gyro0Y std','Gyro0Z std',
                             'Accel0X std', 'Accel0Y std', 'Accel0Z std',
                             'Gyro1X std', 'Gyro1Y std', 'Gyro1Z std', 
                             'Accel1X std', 'Accel1Y std', 'Accel1Z std']]
                         
ents_col = dataset[['Gyro0X ent', 'Gyro0Y ent', 'Gyro0Z ent', 
                             'Accel0X ent', 'Accel0Y ent', 'Accel0Z ent', 
                             'Gyro1X ent', 'Gyro1Y ent','Gyro1Z ent',
                             'Accel1X ent','Accel1Y ent','Accel1Z ent']]
                         
corrs_col = dataset[['Gyro0X corr', 'Gyro0Y corr', 'Gyro0Z corr', 
                             'Accel0X corr', 'Accel0Y corr', 'Accel0Z corr', 
                             'Gyro1X corr', 'Gyro1Y corr','Gyro1Z corr',
                             'Accel1X corr','Accel1Y corr','Accel1Z corr']]

print(stds_col.max())
print(stds_col.min())

means_col = round((means_col+1)*127)
means_col = means_col.astype('uint8')

corrs_col = round((corrs_col+1)*127)
corrs_col = corrs_col.astype('uint8')

stds_col = round((stds_col/max(stds_col.max()))*255)
stds_col = stds_col.astype('uint8')

ents_col = round((ents_col/max(ents_col.max()))*255)
ents_col = ents_col.astype('uint8')

print(stds_col.max())
print(stds_col.min())
```

```python
dataset_NM = (((step_type_col.join(means_col)).join(stds_col)).join(ents_col)).join(corrs_col)
```

```python
#split de la table d'attributs en ensemble de test et d'entraînement
test_nb = 8
X = dataset_NM.iloc[:,1:]
y = dataset_NM.iloc[:,0]
for n in range(1,test_nb+1):
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=n, test_size=0.2)
    dataset_NM_train = (y_train.to_frame()).join(X_train)
    dataset_NM_test = (y_test.to_frame()).join(X_test)
    dataset_NM_train.to_csv("etapdance2/j_score_math_features/NM500/" + dancer + "/train/train"+ str(n) + ".txt",index=False,header=None)
    dataset_NM_test.to_csv("etapdance2/j_score_math_features/NM500/" + dancer + "/test/test"+ str(n) + ".txt",index=False,header=None)
```
