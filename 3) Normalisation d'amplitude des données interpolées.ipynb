```python
#dimensions graphiques
largeur=15
hauteur=5

step_type_array = ["brush_back","brush_forward","dig","heel","step"]
size_array = [29,29,29,29,29]
ext = ".txt"s
dancer = "Daniel"

for i in range(len(size_array)):
    step_type = step_type_array[i]
    size = size_array[i]
    
    path_prep =  "etapdance2" + "/" + dancer + "/" + step_type + "/" + step_type + "_cut_int" + "/" + step_type + "_cut_int_"

    for n in range(size+1):
        step_type_nb = str(n)
        ext = ".txt"
        data_input = path_prep + step_type_nb + ext

        data_int = pandas.read_csv(data_input,names=['Gyro0X (deg)','Gyro0Y (deg)','Gyro0Z (deg)','Accel0X (g)','Accel0Y (g)','Accel0Z (g)','Gyro1X (deg)','Gyro1Y (deg)','Gyro1Z (deg)','Accel1X (g)','Accel1Y (g)','Accel1Z (g)'], dtype='a') 


        # MPU0 6050 (talon)
        Gyro0X = np.array(data_int['Gyro0X (deg)'])[1:].astype(float)
        Gyro0Y = np.array(data_int['Gyro0Y (deg)'])[1:].astype(float)
        Gyro0Z = np.array(data_int['Gyro0Z (deg)'])[1:].astype(float)

        Accel0X = np.array(data_int['Accel0X (g)'])[1:].astype(float)
        Accel0Y = np.array(data_int['Accel0Y (g)'])[1:].astype(float)
        Accel0Z = np.array(data_int['Accel0Z (g)'])[1:].astype(float)

        # MPU1 6050 (pointe)
        Gyro1X = np.array(data_int['Gyro1X (deg)'])[1:].astype(float)
        Gyro1Y = np.array(data_int['Gyro1Y (deg)'])[1:].astype(float)
        Gyro1Z = np.array(data_int['Gyro1Z (deg)'])[1:].astype(float)

        Accel1X = np.array(data_int['Accel1X (g)'])[1:].astype(float)
        Accel1Y = np.array(data_int['Accel1Y (g)'])[1:].astype(float)
        Accel1Z = np.array(data_int['Accel1Z (g)'])[1:].astype(float)

        Gyro0X_prep = Gyro0X/max(abs(Gyro0X))
        Gyro0Y_prep = Gyro0Y/max(abs(Gyro0Y))
        Gyro0Z_prep = Gyro0Z/max(abs(Gyro0Z))
        Accel0X_prep = Accel0X/max(abs(Accel0X))
        Accel0Y_prep = Accel0Y/max(abs(Accel0Y))
        Accel0Z_prep = Accel0Z/max(abs(Accel0Z))

        Gyro1X_prep = Gyro1X/max(abs(Gyro1X))
        Gyro1Y_prep = Gyro1Y/max(abs(Gyro1Y))
        Gyro1Z_prep = Gyro1Z/max(abs(Gyro1Z))
        Accel1X_prep = Accel1X/max(abs(Accel1X))
        Accel1Y_prep = Accel1Y/max(abs(Accel1Y))
        Accel1Z_prep = Accel1Z/max(abs(Accel1Z))

        headers=['Gyro0X (deg)','Gyro0Y (deg)','Gyro0Z (deg)',
             'Accel0X (g)','Accel0Y (g)','Accel0Z (g)',
             'Gyro1X (deg)','Gyro1Y (deg)','Gyro1Z (deg)',
             'Accel1X (g)','Accel1Y (g)','Accel1Z (g)']

        df = pandas.DataFrame({'Gyro0X (deg)': Gyro0X_prep,
                               'Gyro0Y (deg)': Gyro0Y_prep,
                               'Gyro0Z (deg)': Gyro0Z_prep,
                               'Accel0X (g)' : Accel0X_prep,
                               'Accel0Y (g)' : Accel0Y_prep,
                               'Accel0Z (g)' : Accel0Z_prep,
                               'Gyro1X (deg)': Gyro1X_prep,
                               'Gyro1Y (deg)': Gyro1Y_prep,
                               'Gyro1Z (deg)': Gyro1Z_prep,
                               'Accel1X (g)' : Accel1X_prep,
                               'Accel1Y (g)' : Accel1Y_prep,
                               'Accel1Z (g)' : Accel1Z_prep})

        data_output = "etapdance2" + "/" + dancer + "/" + step_type + "/" + step_type + "_cut_prep" + "/" + step_type + "_cut_prep_" + step_type_nb + ext
        df.to_csv(path_or_buf=data_output,index=False)
```
