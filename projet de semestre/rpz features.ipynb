```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 15}) # pour augmenter la taille de caract√®re dans les graphiques
largeur = 15
hauteur = 8
```

```python
def feat_mean(label):
    feat_pos = [0,52,104,152,196,242]
    feat_mean_single = []
    feat_df = pd.read_csv('etapdance2/step_features_table_filt_truecorr_full_final.txt')
    feat_extr = np.array(feat_df[label])[1:].astype(float)
    for n in range(len(feat_pos)-1):
        pouet = feat_extr[feat_pos[n]:feat_pos[n+1]]/(feat_pos[n+1]-feat_pos[n])
        feat_mean_single = np.append(feat_mean_single, np.sum(pouet))
    return feat_mean_single
```

```python
# comparaison gyro mean
single_labels = ["brush_back","brush_forward","dig","heel","step"]

plt.figure(figsize=(largeur,hauteur))
plt.plot(single_labels,feat_mean('Gyro0X mean'),label='Gyro0X mean')
plt.plot(single_labels,feat_mean('Gyro0Y mean'),label='Gyro0Y mean')
plt.plot(single_labels,feat_mean('Gyro0Z mean'),label='Gyro0Z mean')

plt.plot(single_labels,feat_mean('Gyro1X mean'),label='Gyro1X mean')
plt.plot(single_labels,feat_mean('Gyro1Y mean'),label='Gyro1Y mean')
plt.plot(single_labels,feat_mean('Gyro1Z mean'),label='Gyro1Z mean')
plt.ylabel('feature')
plt.xlabel('pas')
plt.autoscale(axis='x',tight=True);
plt.legend()
plt.show()

feat_mean('Gyro0X mean')
```
