```python
%matplotlib inline
import csv
import numpy as np
from scipy.signal import *
from numpy.fft import *
import matplotlib.pyplot as plt
from ipywidgets import * # Pour insérer des widgets (éléments graphiques) dans le notebook
# Attention! Installer scikit-learn pour avoir accès aux algorithmes de classification
# Installer version 0.20 au plus récente (0.22.1) pour avoir accès à la nouvelle fonction tree.plot_tree()
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier # Méthode KNN
from pandas import * # Structures de données
from sklearn.naive_bayes import GaussianNB
from scipy.ndimage import interpolation, median_filter
from scipy.stats import entropy
```

```python
step_type_array = ["brush_back","brush_forward","dig","heel","step"]
size_array = [25,25,23,21,22]


for i in range(len(size_array)):
    step_type = step_type_array[i]
    size = size_array[i]
    
    path_prep =  "etapdance2" + "/" + step_type + "/" + step_type + "_prep" + "/" + step_type + "_prep_"
    
    Gyro0X = 0
    Gyro0Y = 0
    Gyro0Z = 0
    Accel0X = 0
    Accel0Y = 0
    Accel0Z = 0
    
    Gyro1X = 0
    Gyro1Y = 0
    Gyro1Z = 0
    Accel1X = 0
    Accel1Y = 0
    Accel1Z = 0

    for n in range(size+1):
        step_type_nb = str(n)
        ext = ".txt"
        data_input = path_prep + step_type_nb + ext

        data_raw = pandas.read_csv(data_input,names=['Gyro0X (deg)','Gyro0Y (deg)','Gyro0Z (deg)','Accel0X (g)','Accel0Y (g)','Accel0Z (g)','Gyro1X (deg)','Gyro1Y (deg)','Gyro1Z (deg)','Accel1X (g)','Accel1Y (g)','Accel1Z (g)'], dtype='a') 


        # MPU0 6050 (talon)
        Gyro0X = Gyro0X + np.array(data_raw['Gyro0X (deg)'])[1:].astype(float)
        Gyro0Y = Gyro0Y + np.array(data_raw['Gyro0Y (deg)'])[1:].astype(float)
        Gyro0Z = Gyro0Z + np.array(data_raw['Gyro0Z (deg)'])[1:].astype(float)

        Accel0X = Accel0X + np.array(data_raw['Accel0X (g)'])[1:].astype(float)
        Accel0Y = Accel0Y + np.array(data_raw['Accel0Y (g)'])[1:].astype(float)
        Accel0Z = Accel0Z + np.array(data_raw['Accel0Z (g)'])[1:].astype(float)


        # MPU1 6050 (pointe)
        Gyro1X = Gyro1X + np.array(data_raw['Gyro1X (deg)'])[1:].astype(float)
        Gyro1Y = Gyro1Y + np.array(data_raw['Gyro1Y (deg)'])[1:].astype(float)
        Gyro1Z = Gyro1Z + np.array(data_raw['Gyro1Z (deg)'])[1:].astype(float)

        Accel1X = Accel1X + np.array(data_raw['Accel1X (g)'])[1:].astype(float)
        Accel1Y = Accel1Y + np.array(data_raw['Accel1Y (g)'])[1:].astype(float)
        Accel1Z = Accel1Z + np.array(data_raw['Accel1Z (g)'])[1:].astype(float)
            
    Gyro0X_mean = Gyro0X/size
    Gyro0Y_mean = Gyro0Y/size
    Gyro0Z_mean = Gyro0Z/size
    Accel0X_mean = Accel0X/size
    Accel0Y_mean = Accel0Y/size
    Accel0Z_mean = Accel0Z/size
    
    Gyro1X_mean = Gyro1X/size
    Gyro1Y_mean = Gyro1Y/size
    Gyro1Z_mean = Gyro1Z/size
    Accel1X_mean = Accel1X/size
    Accel1Y_mean = Accel1Y/size
    Accel1Z_mean = Accel1Z/size
    
    headers=['Gyro0X (deg)','Gyro0Y (deg)','Gyro0Z (deg)',
         'Accel0X (g)','Accel0Y (g)','Accel0Z (g)',
         'Gyro1X (deg)','Gyro1Y (deg)','Gyro1Z (deg)',
         'Accel1X (g)','Accel1Y (g)','Accel1Z (g)']

    df = pandas.DataFrame({'Gyro0X (deg)': Gyro0X_mean,
                           'Gyro0Y (deg)': Gyro0Y_mean,
                           'Gyro0Z (deg)': Gyro0Z_mean,
                           'Accel0X (g)' : Accel0X_mean,
                           'Accel0Y (g)' : Accel0Y_mean,
                           'Accel0Z (g)' : Accel0Z_mean,
                           'Gyro1X (deg)': Gyro1X_mean,
                           'Gyro1Y (deg)': Gyro1Y_mean,
                           'Gyro1Z (deg)': Gyro1Z_mean,
                           'Accel1X (g)' : Accel1X_mean,
                           'Accel1Y (g)' : Accel1Y_mean,
                           'Accel1Z (g)' : Accel1Z_mean})
                   
    data_output = "etapdance2" + "/" + step_type + "/" + step_type + "_prep_" + "mean" + ext
    df.to_csv(path_or_buf=data_output,index=False)
```
