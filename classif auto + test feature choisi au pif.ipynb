```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 10}) # pour augmenter la taille de caractère dans les graphiques
import seaborn as sns; sns.set()

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import jaccard_score 

from decimal import *

# k Nearest Neighbors Classifier
from sklearn.neighbors import KNeighborsClassifier

# Naive Bayes algorithm Classifier
from sklearn.naive_bayes import GaussianNB

# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

from sklearn import svm

import time
```

```python
#Calcul des scores pro ou débutant
start_time = time.time()
# kNN
# tutorial -> https://www.youtube.com/watch?v=4HKqjENq9OU


dataset = pd.read_csv('etapdance2/dataset_cut_daniel_ilyas_math_features_uint8.txt')
# split dataset
labels = ['brush_back','brush_forward','dig','heel','step']
labels_export = ['Brush (B)','Brush (F)','Dig','Heel','Step']
j_scores_single_ft = pd.DataFrame()

X = dataset.iloc[:,1:]
y = dataset.iloc[:,0]

test_nb = 8

j_scores_df = pd.DataFrame()

j_scores_df['Pas'] = labels_export

for n in range(1,test_nb+1):
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=n, test_size=0.2)
    classifierkNN = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
    #fit Model
    classifierkNN.fit(X_train,y_train)
    y_pred = classifierkNN.predict(X_test)
    #cm = confusion_matrix(y_test,y_pred)
    #plt.figure(figsize = (6,6))
    #sns.set(font_scale=1.3) # agrandi la police des matrices de confusion
    #ax = sns.heatmap(cm.T, square=True, annot=True
    #,fmt='d', cbar=False
    #, xticklabels=labels_export
    #, yticklabels=labels_export)
    #bottom, top = ax.get_ylim()
    #ax.set_ylim(bottom + 0.5, top - 0.5)
    #plt.xlabel('Classes réelles')
    #plt.ylabel('Predictions');
    #plt.savefig(fname='cm_test7_kNN_k3.png')

    j_score = jaccard_score(y_test,y_pred,labels,average=None)
    j_scores_df['T'+ str(n) + ' (%)'] = np.around(j_score*100, decimals=2)

print("--- %s milliseconds ---" % ((time.time() - start_time)*1000))

j_scores_df['Moyenne (%)'] = np.around(j_scores_df.mean(axis=1), decimals=2)

j_scores_df['Moyenne (%)'].mean()
```

```python
#Calcul des scores combinaison des 2 datasets pro et débutant
start_time = time.time()
# kNN
# tutorial -> https://www.youtube.com/watch?v=4HKqjENq9OU

#données normalisées seules
dataset_dan = pd.read_csv('etapdance2/dataset_cut_daniel_data_only_uint8.txt')
dataset_il = pd.read_csv('etapdance2/dataset_cut_ilyas_data_only_uint8.txt')

#descripteurs calculé
#dataset_dan = pd.read_csv('etapdance2/dataset_cut_daniel_math_features.txt')
#dataset_il = pd.read_csv('etapdance2/dataset_cut_ilyas_math_features.txt')

# split dataset
labels = ['brush_back','brush_forward','dig','heel','step']
labels_export = ['Brush (B)','Brush (F)','Dig','Heel','Step']
j_scores_single_ft = pd.DataFrame()

#choisi selon score de chaque attribut testé seul (ils on chacun un score > 26%)
#pouet = [12,14,28,29,30,34,35,38,40,44,42,45,46,48]

X_dan = dataset_dan.iloc[:,1:]
y_dan = dataset_dan.iloc[:,0]

X_il = dataset_il.iloc[:,1:]
y_il = dataset_il.iloc[:,0]

X = X_dan.append(X_il)
y = y_dan.append(y_il)

test_nb = 8

j_scores_df = pd.DataFrame()

j_scores_df['Pas'] = labels_export

for n in range(1,test_nb+1):
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=n, test_size=0.2)
    classifierkNN = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
    #fit Model
    classifierkNN.fit(X_train,y_train)
    y_pred = classifierkNN.predict(X_test)
    j_score = jaccard_score(y_test,y_pred,labels,average=None)
    j_scores_df['T'+ str(n) + ' (%)'] = np.around(j_score*100, decimals=2)

print("--- %s milliseconds ---" % ((time.time() - start_time)*1000))

j_scores_df['Moyenne (%)'] = np.around(j_scores_df.mean(axis=1), decimals=2)

j_scores_df['Moyenne (%)'].mean()
```
